```python
import wandb
import os
import yaml

api = wandb.Api()

saves_dir = os.path.join(os.getcwd(), "experiments")
os.makedirs(saves_dir, exist_ok=True)

user = "alberto-abbo-84-bitstrapped"
project = "ddp-obj-det-v1"

runs = api.runs(f"{user}/{project}")

run_names = [run.name for run in runs]
run_ids = [run.id for run in runs]
for run in runs:
    # Run name and ID
    print(run.name, run.id)
```

```python
run_index = -1
run_name = run_names[run_index]
latest_run = api.run(f"{user}/{project}/{run_ids[run_index]}")
```

```python
latest_run.config
```

```python
history = latest_run.history()
```

```python
history
```

```python
# Save
os.makedirs(os.path.join(saves_dir, run_name), exist_ok=True)
history.to_csv(os.path.join(saves_dir, run_name, f"{run_name}_history.csv"), index=False)
config_dict = dict(latest_run.config)
with open(os.path.join(saves_dir, run_name, f"{run_name}_config.yaml"), "w") as file:
    yaml.dump(config_dict, file, default_flow_style=False)
```
